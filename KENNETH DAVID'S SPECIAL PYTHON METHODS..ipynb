# Python Magic Methods: A Practical Guide

**Understanding Dunder Methods for Object-Oriented Programming**

Magic methods (also called **dunder methods** because they have **double underscores** at the start and end) are special methods that let your custom classes behave like built-in Python types.

They are not called directly. Instead, Python invokes them automatically when you use built-in functions, operators, or language syntax.

## What Are Magic Methods?

- Methods that start and end with double underscores: `__methodname__`
- They teach Python how to apply built-in operations to your custom objects
- They make user-defined classes feel like native Python data types
- They are triggered by syntax, not by direct calls

For example, when you write `len(obj)`, Python internally calls `obj.__len__()`. When you write `obj1 == obj2`, Python calls `obj1.__eq__(obj2)`.

## 1. Object Initialization: `__init__()`

The initializer runs automatically when a new object is created. It sets up the object's starting state.

```python
class Car:
    def __init__(self, brand, model, year, color):
        self.brand = brand
        self.model = model
        self.year = year
        self.color = color
        self.odometer = 0

my_car = Car("Toyota", "Corolla", 2023, "Silver")
print(f"I drive a {my_car.color} {my_car.year} {my_car.brand} {my_car.model}")
print(f"Mileage: {my_car.odometer} km")

fleet = [
    Car("Honda", "Civic", 2022, "Blue"),
    Car("Ford", "Mustang", 2024, "Red"),
    Car("Tesla", "Model 3", 2023, "White")
]
for vehicle in fleet:
    print(f"  - {vehicle.brand} {vehicle.model}")
```

## 2. String Representation: `__str__()` and `__repr__()`

- **`__str__()`**: Informal, human-readable text for end users (used by `print()` and `str()`)
- **`__repr__()`**: Formal, unambiguous text for developers (used by `repr()` and in interactive shells)

```python
class Movie:
    def __init__(self, title, director, year, rating):
        self.title = title
        self.director = director
        self.year = year
        self.rating = rating
    
    def __str__(self):
        return f"{self.title} ({self.year}) - Rated {self.rating}/10"
    
    def __repr__(self):
        return f"Movie('{self.title}', '{self.director}', {self.year}, {self.rating})"

film = Movie("Inception", "Christopher Nolan", 2010, 8.8)
print(f"For users: {str(film)}")
print(f"For devs:  {repr(film)}")

# Without __str__, print() would show something like: <__main__.Movie object at 0x...>
```

## 3. Length: `__len__()`

Defines what `len()` means for your object. Lets custom objects work naturally with the built-in `len()` function.

```python
class Team:
    def __init__(self, name, members):
        self.name = name
        self.members = list(members)
    
    def __len__(self):
        return len(self.members)
    
    def __str__(self):
        return f"{self.name} ({len(self)} members)"

dev_team = Team("Backend Squad", ["Alice", "Bob", "Charlie", "Diana"])
print(f"Team: {dev_team}")
print(f"Headcount: {len(dev_team)}")  # Calls dev_team.__len__() internally

if len(dev_team) > 3:
    print("This is a full-sized team!")
```

## 4. Equality: `__eq__()`

Controls how `==` behaves. By default, Python checks if two variables point to the same memory address. With `__eq__()`, you can compare based on content instead.

```python
class Ticket:
    def __init__(self, ticket_id, event, seat):
        self.ticket_id = ticket_id
        self.event = event
        self.seat = seat
    
    def __eq__(self, other):
        if not isinstance(other, Ticket):
            return False
        return self.ticket_id == other.ticket_id
    
    def __repr__(self):
        return f"Ticket({self.ticket_id}, '{self.event}', '{self.seat}')"

t1 = Ticket(1001, "Concert", "A12")
t2 = Ticket(1001, "Concert", "A12")  # Same ID, different object in memory
t3 = Ticket(1002, "Concert", "B05")

print(f"t1 == t2: {t1 == t2}")  # True (same ticket_id)
print(f"t1 == t3: {t1 == t3}")  # False (different ticket_id)
print(f"Same object in memory: {t1 is t2}")  # False (different objects)
```

## 5. Operator Overloading

Teaches Python how to use `+`, `-`, `*`, and other operators with your custom objects. This makes mathematical operations feel natural.

```python
class Money:
    def __init__(self, amount, currency="USD"):
        self.amount = amount
        self.currency = currency
    
    def __add__(self, other):
        if self.currency != other.currency:
            raise ValueError("Cannot add different currencies")
        return Money(self.amount + other.amount, self.currency)
    
    def __sub__(self, other):
        if self.currency != other.currency:
            raise ValueError("Cannot subtract different currencies")
        return Money(self.amount - other.amount, self.currency)
    
    def __mul__(self, scalar):
        if not isinstance(scalar, (int, float)):
            return NotImplemented
        return Money(self.amount * scalar, self.currency)
    
    def __str__(self):
        return f"{self.currency} {self.amount:.2f}"

wallet = Money(100) + Money(50)
print(f"Wallet total: {wallet}")

discount = wallet - Money(25)
print(f"After discount: {discount}")

triple = Money(10) * 3
print(f"Triple: {triple}")
```

## 6. Index Access: `__getitem__()`

Lets your object support bracket notation `obj[index]`, just like lists and dictionaries.

```python
class Inventory:
    def __init__(self, items):
        self.items = list(items)
    
    def __getitem__(self, index):
        return self.items[index]
    
    def __str__(self):
        return f"Inventory: {self.items}"

store = Inventory(["Laptop", "Monitor", "Keyboard", "Mouse", "Webcam"])
print(store)
print(f"First item: {store[0]}")
print(f"Last item: {store[-1]}")
print(f"Slice [1:3]: {store[1:3]}")  # Slicing works too!
```

## 7. Index Assignment: `__setitem__()`

Enables assignment through bracket notation: `obj[index] = value`. Useful for building custom container classes.

```python
class Inventory:
    def __init__(self, items):
        self.items = list(items)
    
    def __getitem__(self, index):
        return self.items[index]
    
    def __setitem__(self, index, value):
        self.items[index] = value
    
    def __str__(self):
        return f"Inventory: {self.items}"

store = Inventory(["Laptop", "Monitor", "Keyboard"])
print(f"Before: {store}")

store[1] = "Curved Monitor"  # Calls store.__setitem__(1, "Curved Monitor")
print(f"After:  {store}")

store[0] = "Gaming Laptop"
print(f"Final:  {store}")
```

## 8. Membership Testing: `__contains__()`

Defines how the `in` operator works for your object. Lets you check if an item belongs to your collection.

```python
class Classroom:
    def __init__(self, room_name, students):
        self.room_name = room_name
        self.students = list(students)
    
    def __contains__(self, student):
        return student in self.students
    
    def __len__(self):
        return len(self.students)
    
    def __str__(self):
        return f"{self.room_name}: {len(self)} students"

cs101 = Classroom("CS-101", ["Emma", "Liam", "Olivia", "Noah", "Ava"])
print(cs101)

print(f"Is Emma enrolled? {'Emma' in cs101}")
print(f"Is Jack enrolled? {'Jack' in cs101}")

if 'Olivia' in cs101:
    print("Olivia is present today.")
```

## 9. Callable Objects: `__call__()`

Makes an object behave like a function. You can invoke it with parentheses while it retains internal state.

```python
class TaxCalculator:
    def __init__(self, rate):
        self.rate = rate
    
    def __call__(self, amount):
        return amount * self.rate
    
    def __str__(self):
        return f"Tax Calculator ({self.rate*100:.1f}%)"

vat = TaxCalculator(0.20)   # 20% tax
sales_tax = TaxCalculator(0.08)  # 8% tax

print(f"{vat}")
print(f"$100 with VAT: ${vat(100):.2f}")
print(f"$250 with VAT: ${vat(250):.2f}")

print(f"\n{sales_tax}")
print(f"$100 with sales tax: ${sales_tax(100):.2f}")

# The object behaves like a function!
print(f"Type check: {type(vat)}")
```

## 10. Iteration: `__iter__()` and `__next__()`

These two methods form the iterator protocol that powers `for` loops.

- **`__iter__()`**: Returns the iterator object (usually `self`)
- **`__next__()`**: Returns the next value or raises `StopIteration` when done

```python
class Fibonacci:
    def __init__(self, max_count):
        self.max_count = max_count
        self.count = 0
        self.a, self.b = 0, 1
    
    def __iter__(self):
        self.count = 0
        self.a, self.b = 0, 1
        return self
    
    def __next__(self):
        if self.count >= self.max_count:
            raise StopIteration
        value = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return value

print("First 10 Fibonacci numbers:")
for num in Fibonacci(10):
    print(num, end=" ")
print()

print("\nFirst 5 Fibonacci numbers:")
fib = Fibonacci(5)
print(list(fib))
```

## 11. Boolean Evaluation: `__bool__()`

Controls how your object behaves in truth-value tests: `if obj:`, `while obj:`, `bool(obj)`, etc.

```python
class Stock:
    def __init__(self, product, quantity):
        self.product = product
        self.quantity = quantity
    
    def __bool__(self):
        return self.quantity > 0
    
    def __str__(self):
        status = "In Stock" if self else "Out of Stock"
        return f"{self.product}: {status} ({self.quantity} units)"

item1 = Stock("Wireless Mouse", 15)
item2 = Stock("Mechanical Keyboard", 0)
item3 = Stock("USB-C Cable", 3)

for item in [item1, item2, item3]:
    print(item)

print("\nAvailable for purchase:")
for item in [item1, item2, item3]:
    if item:
        print(f"  - {item.product}")
```

## 12. Cleanup: `__del__()`

Called when an object is about to be destroyed. Historically used for cleanup, but modern Python prefers **context managers** (`with` statements) because garbage collection timing is unpredictable.

**Note:** Do not rely on `__del__` for critical cleanup operations.

```python
class TempFile:
    def __init__(self, filename):
        self.filename = filename
        print(f"Created temporary file: {filename}")
    
    def __del__(self):
        print(f"Temporary file '{self.filename}' marked for cleanup")

log = TempFile("debug.log")
print("Processing data...")
del log  # Triggers __del__
print("Done.")

print("\n--- Another example ---")
class NetworkConnection:
    def __init__(self, host):
        self.host = host
        print(f"Connected to {host}")
    
    def __del__(self):
        print(f"Disconnected from {self.host}")

conn = NetworkConnection("api.example.com")
del conn
```

| 15 | `__bool__` | Truth-value testing | `bool()`, `if` |
| 16 | `__del__` | Destructor/cleanup | `del`, GC |

## Solution

Click below to reveal the solution after attempting the exercise.

```python
class TaskManager:
    def __init__(self, tasks):
        self.tasks = list(tasks)
    
    def __str__(self):
        return f"Tasks: {self.tasks}"
    
    def __len__(self):
        return len(self.tasks)
    
    def __getitem__(self, index):
        return self.tasks[index]
    
    def __setitem__(self, index, value):
        self.tasks[index] = value
    
    def __eq__(self, other):
        if not isinstance(other, TaskManager):
            return False
        return self.tasks == other.tasks
    
    def __add__(self, other):
        if isinstance(other, TaskManager):
            return TaskManager(self.tasks + other.tasks)
        return NotImplemented
    
    def __contains__(self, task):
        return task in self.tasks
    
    def __bool__(self):
        return len(self.tasks) > 0


# Verification
tm1 = TaskManager(["Code review", "Write tests", "Deploy"])
tm2 = TaskManager(["Code review", "Write tests", "Deploy"])

print(tm1)
print(f"Count: {len(tm1)}")
print(f"First: {tm1[0]}")

tm1[1] = "Update docs"
print(tm1)
print(f"Equal? {tm1 == tm2}")

tm3 = tm1 + tm2
print(tm3)

print(f"Has 'Deploy'? {'Deploy' in tm1}")
print(f"Has tasks? {bool(tm1)}")

empty = TaskManager([])
print(f"Empty has tasks? {bool(empty)}")
```

# Key Takeaways

1. **Magic methods** bridge the gap between custom classes and Python's built-in operations.
2. They are invoked **automatically** by Python syntax, not called directly.
3. **`__str__`** is for users; **`__repr__`** is for developers and debugging.
4. **`__eq__`** lets you compare objects by content rather than memory address.
5. **Operator overloading** (`__add__`, `__sub__`, `__mul__`) makes custom types intuitive to use.
6. **`__getitem__`** and **`__setitem__`** bring the power of indexing to your objects.
7. **`__iter__`** and **`__next__`** let your objects work in `for` loops.
8. **`__call__`** turns objects into functions that remember state.
9. **`__bool__`** defines what counts as `True` or `False` for your type.
10. Prefer **context managers** over **`__del__`** for reliable resource cleanup.

---

**Golden Rule:** If Python has a built-in way to interact with data (printing, adding, comparing, iterating), there is probably a magic method that lets your class participate.
